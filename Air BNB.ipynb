```python
# Drop duplicates record

df.drop_duplicates(inplace=True)

#drop house rules and license columns with insufficient data
df.drop(['house_rules','license'],axis=1, inplace= True)

#Remove all dollar signs in the price and service fee columns
df['price'] = df['price'].astype(str).str.strip().str.replace('$', '', regex=False).astype(float)
df['service fee']=df['service fee'].astype(str).str.strip().str.replace('$','',regex=False)

#Remove all columns in the price and service fee columns
df['price']=df['price'].astype(str).str.strip().str.replace(',','',regex=False)
df['service fee']=df['service fee'].str.replace(',','',regex=False)

#Remove the price and service fee columns to include a dollar sign
df.rename(columns={
    'price':'price_$',
    'service fee':'service_fee_$'
},inplace=True)

#drop all records with missing values
df.dropna(inplace=True)

#change all mismatched data types to the appropriate once.
df['price_$']=df['price_$'].astype(float)
df['service_fee_$']=df['service_fee_$'].astype(float)
df['id']=df['id'].astype(str)
df['host id']=df['host id'].astype(str)
df['last review']=pd.to_datetime(df['last review'])
df['Construction year']=df['Construction year'].astype(int)

#correct the spelling of 'brooklyn to 'Brooklyn
df.loc[df['neighbourhood group'] == 'brooklyn','neighbourhood group']= 'Brooklyn'
df.loc[df['neighbourhood group'] == 'brookln','neighbourhood group']= 'Brooklyn'


#get rid of outliers in the 'availability 365' column data
df=df.drop(df[df['availability 365']>500].index)
```

```python
# Assuming your room type column is named 'room type' or 'room_type' after cleaning.
# We will use a try/except block to handle either column name.

try:
    # 1. Calculate value counts and ensure the index (room types) are simple strings
    property_types_raw = df['room type'].astype(str).value_counts()
except KeyError:
    # Try the alternate name
    property_types_raw = df['room_type'].astype(str).value_counts()

# 2. Extract and convert the data into standard Python lists
# X-axis data (labels)
x_data = property_types_raw.index.tolist() 
# Y-axis data (counts), converted to float to handle plotting cleanly
y_data = property_types_raw.values.astype(float).tolist() 

print("Data successfully cleaned for plotting. X-axis items:", len(x_data))
print("Y-axis items:", len(y_data))
```

```python
# Room type count plotted on a barchart
plt.figure(figsize=(10, 6))

# Use the cleaned x_data (Room Types) and y_data (Counts)
room_type_bar = plt.bar(x_data, y_data); 

# Add labels using y_data (the counts)
plt.bar_label(room_type_bar, labels=y_data, padding=4);

# Set plot limits, labels, and title
plt.ylim([0, 50000]);
plt.xlabel('Room Type');
plt.ylabel('Room Type Count');
plt.title('Property Types and their Count in the Dataset');

# Rotate x-labels for readability if you have many types
plt.xticks(rotation=45, ha='right')
plt.tight_layout() # Adjusts layout to prevent labels from being cut off
plt.show()
```

# Which neighborhoods group has the highest average prices for AirBnb listings?

```python
# Which neighborhoods group has the highest average prices for AirBnb listings?
avg_price=df.groupby('neighbourhood group')['price_$'].mean().sort_values(ascending=False).to_frame()
#avg_price_bar=plt.bar(avg_price,avg_price.loc[:,'price_$']);
#plt.bar_label(avg_price_bar,labels=round(avg_price.loc[:"price_$"],2),label_type='edge',padding=4);
#plt.ylim([0,700]);
#plt.xlabel('Neighborhood Groups');
#plt.ylabel('Average price per listing ($)');
#plt.title('Average price per listing ($) in each neighborhood group ');


# Extract data for plotting
x_data = avg_price.index.tolist()
y_data = avg_price['price_$'].values.tolist()

# --- Plotting ---
plt.figure(figsize=(10, 6))

# Use the Index (Neighborhood Groups) for X and the column values (Average Price) for Y
avg_price_bar = plt.bar(x_data, y_data, color='#9c3434')

# Format labels to include a dollar sign and comma separation
labels = [f'${p:,.2f}' for p in y_data]
plt.bar_label(avg_price_bar, labels=labels, padding=4)

# Set Y-limit slightly above the max price for label clarity
plt.ylim([0, max(y_data) * 1.15])

# Set labels and title
plt.xlabel('Neighborhood Groups')
plt.ylabel('Average Price ($)')
plt.title('Neighbourhood Group Average Listing Prices')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\n--- Average Price per Neighbourhood Group ---")
print(avg_price)
```

Positive Correlation: As the number of listings a host has increases, their listings tend to be slightly more available throughout the year.

Weak Magnitude (0.135): The relationship is not strong. While there is a tendency, the number of listings a host manages is not a major factor in determining the overall availability of their properties. Many other factors influence a listing's availability more significantly.
